```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df["math score"], label="Math")
sns.kdeplot(df["reading score"], label="Reading")
sns.kdeplot(df["writing score"], label="Writing")
plt.title("Score Distributions Across Subjects")
plt.xlabel("Score")
plt.legend()
plt.show()
```

```python
sns.boxplot(data=df[["math score","reading score","writing score"]])
plt.title("Outliers Across Subjects")
plt.ylabel("Score")
plt.show()
```

```python
sns.heatmap(df[["math score","reading score","writing score"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Subject Scores")
plt.show()
```

```python
df_melt = df.melt(id_vars="test preparation course",value_vars=["math score","reading score","writing score"],var_name="Subject",value_name="Score")
sns.boxplot(x="test preparation course", y="Score", hue="Subject", data=df_melt)
plt.title("Scores by Test Preparation Course")
plt.show()
```

```python
df_melt = df.melt(id_vars="lunch",value_vars=["math score","reading score","writing score"],var_name="Subject",value_name="Score")

sns.boxplot(x="lunch", y="Score", hue="Subject", data=df_melt)
plt.title("Scores by Lunch Type")
plt.show()
```
